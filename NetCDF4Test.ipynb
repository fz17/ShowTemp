```python
dir = './Samples/lnd/'
nc_temp_list = list()

for file in os.listdir(dir):
    if file.split('.')[-1] == 'nc':
        print(file)
        nc_temp_list.append('{}{}'.format(dir, file))
print()

for i in np.arange(len(nc_temp_list)):
    print(nc_temp_list[i])
```

<center> <font color = #82318E size = 3 face = "微软雅黑" > 1.3.2 Variables </font> </center>

```python
temp_pointer = nc.Dataset(nc_temp_list[0])

temp_dict = {}

for j in np.arange(len(list(temp_pointer.variables.keys()))):
    temp = temp_pointer.variables[list(temp_pointer.variables.keys())[j]]
         
    print('-'*25 + ' Variable {0:03d} '.format(j) + '-'*25)
    print(temp.name)
      
    if temp.name != 'time' and temp.name != 'mcdate' and temp.name != 'mcsec' and temp.name != 'mdcur' and \
    temp.name != 'mscur' and temp.name != 'nstep' and temp.name != 'time_bounds' and temp.name != 'date_written' and temp.name != 'time_written':
        print(temp.long_name)
    elif temp.name != 'date_written' and temp.name != 'time_written':
        print(temp.logname)
    else:
        print(temp.name)
        
    print(temp.shape)
    print('-'*64, end = '\n'*2)
    
    if temp.name == 'time' or temp.name == 'mcdate' or temp.name == 'mcsec' or temp.name == 'mdcur' or \
    temp.name == 'mscur' or temp.name == 'nstep' or temp.name == 'time_bounds':
        temp_dict.setdefault('{0:03d}'.format(j), [temp.name, temp.logname, temp.shape])        
    elif temp.name == 'date_written' or temp.name == 'time_written':
        temp_dict.setdefault('{0:03d}'.format(j), [temp.name, temp.name, temp.shape])
    else:
        temp_dict.setdefault('{0:03d}'.format(j), [temp.name, temp.long_name, temp.shape])
    
temp_df = pd.DataFrame(temp_dict, index = ['Name', 'Long_name', 'Shape']).transpose()
print(temp_df)

temp_df.to_excel('{}lnd_Variables.xlsx'.format(dir), sheet_name = 'Sheet3')
```

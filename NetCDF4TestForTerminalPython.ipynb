```python
dir = './Samples/atm/'
nc_temp_list = list()

for file in os.listdir(dir):
    if file.split('.')[-1] == 'nc':
        print(file)
        nc_temp_list.append('{}{}'.format(dir, file))
print()

for i in np.arange(len(nc_temp_list)):
    print(nc_temp_list[i])
```

```python
temp_pointer = nc.Dataset(nc_temp_list[0])

temp_dict = {}

for j in np.arange(len(list(temp_pointer.variables.keys()))):
    temp = temp_pointer.variables[list(temp_pointer.variables.keys())[j]]

    '''
    print('-'*25 + ' Variable {0:03d} '.format(j) + '-'*25)
    print(temp)
    print('-'*64, end = '\n'*2)
    '''
    
    if temp.name != 'date_written' and temp.name != 'time_written':
        temp_dict.setdefault('{0:03d}'.format(j), [temp.name, temp.long_name, temp.shape])
    else:
        temp_dict.setdefault('{0:03d}'.format(j), [temp.name, temp.name, temp.shape])
    
temp_df = pd.DataFrame(temp_dict, index = ['Name', 'Long_name', 'Shape']).transpose()
print(temp_df)

temp_df.to_excel('{}atm_Variables.xlsx'.format(dir), sheet_name = 'Sheet1')
```

```python
def NC2TimeSeriesNPY(*, string, temp_shape):

    I = len(temp_list)
    if temp_shape == 3:
        J = nc.Dataset(temp_list[0]).variables[string].shape[-3]
        K = nc.Dataset(temp_list[0]).variables[string].shape[-2]
        L = nc.Dataset(temp_list[0]).variables[string].shape[-1]
        temp_matrix = np.ones((I, J, K, L))
        print(temp_matrix.shape)

        for i in np.arange(I):
            with nc.Dataset(temp_list[i]) as temp_pointer:
                temp_single_matrix = temp_pointer.variables[string]
                print(temp_single_matrix.shape)
                temp_matrix[i, ::, ::, ::] = temp_single_matrix[::]


    elif temp_shape == 2:
        J = nc.Dataset(temp_list[0]).variables[string].shape[-2]
        K = nc.Dataset(temp_list[0]).variables[string].shape[-1]
        temp_matrix = np.ones((I, J, K))
        print(temp_matrix.shape)

        for i in np.arange(I):
            with nc.Dataset(temp_list[i]) as temp_pointer:
                temp_single_matrix = temp_pointer.variables[string]
                print(temp_single_matrix.shape)
                temp_matrix[i, ::, ::] = temp_single_matrix[::]

    elif temp_shape == 1:
        J = nc.Dataset(temp_list[0]).variables[string].shape[-1]
        temp_matrix = np.ones((I, J))
        print(temp_matrix.shape)

        for i in np.arange(I):
            with nc.Dataset(temp_list[i]) as temp_pointer:
                temp_single_matrix = temp_pointer.variables[string]
                print(temp_single_matrix.shape)
                temp_matrix[i, ::] = temp_single_matrix[::]

    else:
        temp_matrix = np.ones((I, ))
        print(temp_matrix.shape)

        for i in np.arange(I):
            with nc.Dataset(temp_list[i]) as temp_pointer:
                temp_single_matrix = temp_pointer.variables[string]
                print(temp_single_matrix.shape)
                temp_matrix[i] = temp_single_matrix[::]

    np.save('./Tensor/atm/{}.npy'.format(string), temp_matrix)        
```
